```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.newschannel10.com/"
```

```python
page=requests.get(url)
```

```python
page
```

```python
page.status_code
```

```python
page.text
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
soup.prettify()
```

```python
title=[]
date=[]
author=[]
```

```python
for i in soup.findAll('a',attrs={'class':'unstyled-link'}):
    title.append(i.text)
title
```

```python
len(title)
```

```python
for j in soup.findAll('span',attrs={'class':'hidden-sm'}):
    date.append(j.text)
date
```

```python
len(date)
```

```python
for k in soup.findAll('h3',attrs={'class':'byline bold'}):
    author.append(k.text)
author
```

```python
len(author)
```

```python
import pandas as pd
```

```python
data={'Title':title}
df=pd.DataFrame(data)
df.reset_index(level=0,inplace=True)
df
```
